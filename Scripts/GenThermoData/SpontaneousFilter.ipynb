```python
import pandas as pd
import numpy as np
```

```python
def spontaneous_filter(Network, Generation):
    df = pd.read_csv(f'./RelsDataWithThermo/{Network}/{Network}_{Generation}RelsWithThermo.tsv', sep='\t')
    Indexes = []
    Reagents = []
    Products = []
    Rules = []
    Energy_Changes = []
    for i in range(len(df['Index'])):
        if df['Energy Change'][i] <= 0:
            Indexes.append(df['Index'][i])
            Reagents.append(df['Reagents'][i])
            Products.append(df['Products'][i])
            Rules.append(df['Rule'][i])
            Energy_Changes.append(df['Energy Change'][i])
    data = {'Index':Indexes, 'Reagents':Reagents, 'Products':Products, 'Rule':Rules, 'Energy Change':Energy_Changes}
    output = pd.DataFrame(data)
    output.to_csv(f'./SpontaneousRelsDataWithThermo/{Network}/Spontaneous{Network}_{Generation}RelsWithThermo.tsv', header=None, index=None, sep='\t', mode='a')
    return(output)
```

```python
%%time
spontaneous_filter('Formose', '2')
```
